```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
# 1 data preprocessing
# 2 Model selection
# 3 Model fit
# 4 Model predict
```

```python
x = np.linspace(-80,80,num=200)
```

```python
x
```

```python
plt.plot(x)
```

```python
noise = np.random.randint(-80,80,size=(x.shape[0]))
```

```python
x.shape
```

```python
y = 0.6*x + noise
```

```python
plt.plot(y)
```

```python
# select the Model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression()
```

```python
# fit the model
model.fit(x[:,np.newaxis],y[:,np.newaxis])
```

```python
m, c= model.coef_ , model.intercept_
```

```python
m , c
```

```python
plt.scatter(x,y)
plt.plot(x,m*x[:,np.newaxis] + c,c='r')
plt.plot(x,1.8*x[:,np.newaxis] + c,c='g')
```

```python
model.predict([[42]])
```

```python
test_x = np.random.randint(80, size=(x.shape))
```

```python
test_x
```

```python
model.predict(test_x[:,np.newaxis])
```
